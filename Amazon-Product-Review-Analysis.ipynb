```python
import matplotlib.pyplot as plt
import seaborn as sns

# General graphic design
sns.set(style="whitegrid", palette="muted", font_scale=1.2)
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Rating distribution graph
plt.figure(figsize=(10, 6))

sns.countplot(x='reviews.rating', data=df, palette='viridis')

plt.title('Distribution of Product Ratings', fontsize=18)

plt.xlabel('Rating', fontsize=14)
plt.ylabel('Number of Reviews', fontsize=14)

for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha = 'center', va = 'center', fontsize=12, color='black', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()

plt.show()
```

```python
#Average rating by product
top_products = df.groupby('name')['reviews.rating'].agg(['count', 'mean'])
top_products = top_products[top_products['count'] > 50].sort_values(by='mean', ascending=False).head(10)

sns.barplot(x=top_products['mean'], y=top_products.index, palette='viridis')

plt.title('Top 10 Rated Products (50+ Reviews)', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

# Merge all reviews into one string
text = ' '.join(str(review) for review in df['reviews.text'].dropna())

wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='white',
                      width=800,
                      height=400).generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Reviews', fontsize=16)
plt.show()
```

```python
df_exploded = df.explode('primaryCategories')

print(df_exploded['primaryCategories'].head(10))
```

```python
# Create a rating distribution graph by categories
plt.figure(figsize=(12, 6))

sns.countplot(x='primaryCategories', hue='reviews.rating', data=df_exploded, palette='Blues')

plt.title('Distribution of Ratings by Product Category', fontsize=18)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Number of Reviews', fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Data filtering - focus on reviews that received more than 0 helpful votes
df_filtered = df[df['reviews.numHelpful'] > 0]

plt.figure(figsize=(12, 6))

# Create a bar plot of the number of helpful views
sns.countplot(x='reviews.numHelpful', data=df_filtered, palette='Blues')

plt.title('Distribution of Helpful Votes on Reviews (Excluding 0 votes)', fontsize=18)
plt.xlabel('Number of Helpful Votes', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
df['reviews.date'] = pd.to_datetime(df['reviews.date'], errors='coerce')
monthly_reviews = df.groupby(df['reviews.date'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
monthly_reviews.plot(kind='line', marker='o', color='teal')
plt.title('Number of Reviews Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by rating and whether they recommend
grouped = df.groupby(['reviews.rating', 'reviews.doRecommend']).size().unstack()

# Only existing ratings
ratings = sorted(grouped.index)

# Create a pie chart for each rating
fig, axs = plt.subplots(1, len(ratings), figsize=(22, 6))
fig.suptitle('Recommendation Breakdown by Rating', fontsize=20)

for i, rating in enumerate(ratings):
    if rating in grouped.index:
        axs[i].pie(
            grouped.loc[rating],
            labels=['Not Recommended', 'Recommended'],
            autopct='%1.1f%%',
            startangle=90,
            colors=['#FF6B6B', '#4CAF50'],
            textprops={'fontsize': 12}
        )
        axs[i].set_title(f'Rating {rating}', fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns
corr_data = df[['reviews.rating', 'reviews.numHelpful']]

# Compute correlation matrix
correlation_matrix = corr_data.corr()

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title and explanation
plt.title('Correlation Between Rating and Number of Helpful Votes', fontsize=14, pad=20)

# Add explanation text below the plot with extra padding
plt.figtext(0.5, -0.2,  # Adjust the y-position for more space
    'This heatmap shows the correlation between product ratings and the number of helpful votes.\n'
    'A value close to 1 indicates a strong positive relationship (higher ratings get more helpful votes).\n'
    'A value near 0 means little or no relationship, while negative values indicate an inverse relationship.',
    wrap=True, horizontalalignment='center', fontsize=10)

plt.tight_layout()
plt.show()
```
